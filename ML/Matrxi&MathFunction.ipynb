## Exercise 1: Matrix Operations Program


```python
import numpy as np


def read_int(prompt, minimum=None, maximum=None):
    """Read an integer from the user with optional range validation."""
    while True:
        try:
            value = int(input(prompt))

            if minimum is not None and value < minimum:
                print(f"Please enter a value greater than or equal to {minimum}.")
                continue

            if maximum is not None and value > maximum:
                print(f"Please enter a value less than or equal to {maximum}.")
                continue

            return value
        except ValueError:
            print("Invalid input. Please enter a whole number.")


def read_float(prompt):
    """Read a numeric value from the user."""
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Invalid input. Please enter a numeric value.")


def read_matrix(name):
    """Create a NumPy matrix by taking row, column, and element values."""
    print(f"\nEnter details for Matrix {name}")
    rows = read_int("Number of rows: ", minimum=1)
    cols = read_int("Number of columns: ", minimum=1)

    elements = []

    # Read each matrix element from the user.
    for i in range(rows):
        row = []
        for j in range(cols):
            value = read_float(f"Enter element [{i + 1}][{j + 1}]: ")
            row.append(value)
        elements.append(row)

    return np.array(elements)


def display_matrix(title, matrix):
    """Display a matrix with a heading."""
    print(f"\n{title}")
    print("-" * len(title))
    print(matrix)


def matrix_addition():
    """Add two matrices after checking their dimensions."""
    matrix_a = read_matrix("A")
    matrix_b = read_matrix("B")

    if matrix_a.shape != matrix_b.shape:
        print("\nError: Addition requires both matrices to have the same dimensions.")
        return

    display_matrix("Matrix A + Matrix B", matrix_a + matrix_b)


def matrix_subtraction():
    """Subtract two matrices after checking their dimensions."""
    matrix_a = read_matrix("A")
    matrix_b = read_matrix("B")

    if matrix_a.shape != matrix_b.shape:
        print("\nError: Subtraction requires both matrices to have the same dimensions.")
        return

    display_matrix("Matrix A - Matrix B", matrix_a - matrix_b)


def matrix_multiplication():
    """Multiply two matrices after checking multiplication compatibility."""
    matrix_a = read_matrix("A")
    matrix_b = read_matrix("B")

    if matrix_a.shape[1] != matrix_b.shape[0]:
        print("\nError: Columns of Matrix A must be equal to rows of Matrix B.")
        return

    display_matrix("Matrix A x Matrix B", np.matmul(matrix_a, matrix_b))


def matrix_transpose():
    """Find the transpose of a matrix."""
    matrix = read_matrix("A")
    display_matrix("Transpose of Matrix A", np.transpose(matrix))


def matrix_determinant():
    """Find the determinant of a square matrix."""
    matrix = read_matrix("A")

    if matrix.shape[0] != matrix.shape[1]:
        print("\nError: Determinant can be calculated only for a square matrix.")
        return

    determinant = np.linalg.det(matrix)
    print("\nDeterminant")
    print("-----------")
    print(f"det(A) = {determinant:.4f}")


def matrix_min_max():
    """Find the minimum and maximum values in a matrix."""
    matrix = read_matrix("A")
    print("\nMinimum and Maximum Values")
    print("--------------------------")
    print(f"Minimum value: {np.min(matrix)}")
    print(f"Maximum value: {np.max(matrix)}")


def matrix_diagonal_values():
    """Display the main diagonal values of a matrix."""
    matrix = read_matrix("A")
    diagonal = np.diagonal(matrix)

    print("\nDiagonal Values")
    print("---------------")
    print(diagonal)


def matrix_row_column_sum():
    """Display row-wise and column-wise sum of a matrix."""
    matrix = read_matrix("A")
    row_sum = np.sum(matrix, axis=1)
    column_sum = np.sum(matrix, axis=0)

    print("\nRow and Column Sum")
    print("------------------")
    print(f"Row-wise sum: {row_sum}")
    print(f"Column-wise sum: {column_sum}")


def matrix_subarray():
    """Extract and display a subarray from a matrix using row and column ranges."""
    matrix = read_matrix("A")
    rows, cols = matrix.shape

    print("\nEnter subarray range using 1-based positions.")
    start_row = read_int("Start row: ", minimum=1, maximum=rows)
    end_row = read_int("End row: ", minimum=start_row, maximum=rows)
    start_col = read_int("Start column: ", minimum=1, maximum=cols)
    end_col = read_int("End column: ", minimum=start_col, maximum=cols)

    # Convert 1-based user input to Python's 0-based slicing.
    subarray = matrix[start_row - 1:end_row, start_col - 1:end_col]
    display_matrix("Selected Subarray", subarray)


def matrix_operations_menu():
    """Display the matrix operations menu repeatedly."""
    while True:
        print("\n========== Matrix Operations ==========")
        print("1. Matrix Addition")
        print("2. Matrix Subtraction")
        print("3. Matrix Multiplication")
        print("4. Matrix Transpose")
        print("5. Determinant")
        print("6. Min Max")
        print("7. Diagonal Values")
        print("8. Row and Column Sum")
        print("9. Subarray")
        print("10. Exit Matrix Program")

        choice = input("Enter your choice: ").strip()

        # Match-case handles user choices in Python 3.10+.
        match choice:
            case "1":
                matrix_addition()
            case "2":
                matrix_subtraction()
            case "3":
                matrix_multiplication()
            case "4":
                matrix_transpose()
            case "5":
                matrix_determinant()
            case "6":
                matrix_min_max()
            case "7":
                matrix_diagonal_values()
            case "8":
                matrix_row_column_sum()
            case "9":
                matrix_subarray()
            case "10":
                print("\nExiting Matrix Operations Program.")
                break
            case _:
                print("Invalid choice. Please select a valid matrix operation.")


# Start the matrix program.
matrix_operations_menu()
```

## Exercise 2: Mathematical Functions Program


```python
import math


def read_int(prompt, minimum=None):
    """Read an integer from the user with optional minimum validation."""
    while True:
        try:
            value = int(input(prompt))

            if minimum is not None and value < minimum:
                print(f"Please enter a value greater than or equal to {minimum}.")
                continue

            return value
        except ValueError:
            print("Invalid input. Please enter a whole number.")


def read_float(prompt):
    """Read a numeric value from the user."""
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Invalid input. Please enter a numeric value.")


def square_root():
    """Calculate square root using math.sqrt."""
    number = read_float("Enter a non-negative number: ")

    if number < 0:
        print("\nError: Square root is not defined for negative numbers in real numbers.")
        return

    print(f"\nSquare root of {number} = {math.sqrt(number):.4f}")


def power():
    """Calculate power using math.pow."""
    base = read_float("Enter base: ")
    exponent = read_float("Enter exponent: ")
    print(f"\n{base} raised to {exponent} = {math.pow(base, exponent):.4f}")


def factorial():
    """Calculate factorial using math.factorial."""
    number = read_int("Enter a non-negative integer: ", minimum=0)
    print(f"\nFactorial of {number} = {math.factorial(number)}")


def logarithm():
    """Calculate logarithm using math.log."""
    number = read_float("Enter a positive number: ")

    if number <= 0:
        print("\nError: Logarithm requires a positive number.")
        return

    print("Choose logarithm type:")
    print("1. Natural Logarithm")
    print("2. Logarithm with Custom Base")
    choice = input("Enter your choice: ").strip()

    match choice:
        case "1":
            print(f"\nln({number}) = {math.log(number):.4f}")
        case "2":
            base = read_float("Enter base greater than 0 and not equal to 1: ")
            if base <= 0 or base == 1:
                print("\nError: Base must be greater than 0 and not equal to 1.")
                return
            print(f"\nlog base {base} of {number} = {math.log(number, base):.4f}")
        case _:
            print("Invalid choice. Returning to mathematical functions menu.")


def sine():
    """Calculate sine of an angle in radians."""
    angle = read_float("Enter angle in radians: ")
    print(f"\nsin({angle}) = {math.sin(angle):.4f}")


def cosine():
    """Calculate cosine of an angle in radians."""
    angle = read_float("Enter angle in radians: ")
    print(f"\ncos({angle}) = {math.cos(angle):.4f}")


def tangent():
    """Calculate tangent of an angle in radians."""
    angle = read_float("Enter angle in radians: ")
    print(f"\ntan({angle}) = {math.tan(angle):.4f}")


def degree_to_radian():
    """Convert degrees to radians using math.radians."""
    degrees = read_float("Enter angle in degrees: ")
    print(f"\n{degrees} degrees = {math.radians(degrees):.4f} radians")


def radian_to_degree():
    """Convert radians to degrees using math.degrees."""
    radians = read_float("Enter angle in radians: ")
    print(f"\n{radians} radians = {math.degrees(radians):.4f} degrees")


def mathematical_functions_menu():
    """Display the mathematical functions menu repeatedly."""
    while True:
        print("\n========== Mathematical Functions ==========")
        print("1. Square Root")
        print("2. Power")
        print("3. Factorial")
        print("4. Logarithm")
        print("5. Sine")
        print("6. Cosine")
        print("7. Tangent")
        print("8. Degree to Radian Conversion")
        print("9. Radian to Degree Conversion")
        print("10. Exit Math Program")

        choice = input("Enter your choice: ").strip()

        # Match-case handles user choices in Python 3.10+.
        match choice:
            case "1":
                square_root()
            case "2":
                power()
            case "3":
                factorial()
            case "4":
                logarithm()
            case "5":
                sine()
            case "6":
                cosine()
            case "7":
                tangent()
            case "8":
                degree_to_radian()
            case "9":
                radian_to_degree()
            case "10":
                print("\nExiting Mathematical Functions Program.")
                break
            case _:
                print("Invalid choice. Please select a valid mathematical function.")


# Start the math program.
mathematical_functions_menu()
```
